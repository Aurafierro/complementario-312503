```python
# Importar bibliotecas necesarias
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo
file_path = 'ImpactoInfraestructuraDesarroll deSoftware.csv'
data = pd.read_csv(file_path, encoding='latin1')

# Limpiar y convertir las columnas seleccionadas
data['Adopcionn (%)'] = data['Adopcionn (%)'].str.replace('%', '').astype(float)
data['Crecimiento Estimado (%)'] = data['Crecimiento Estimado (%)'].str.replace('%', '').astype(float)

# Crear el gráfico de correlación
plt.figure(figsize=(8, 6))
plt.scatter(data['Adopcionn (%)'], data['Crecimiento Estimado (%)'], alpha=0.7)
plt.title('Correlación entre Adopción y Crecimiento Estimado')
plt.xlabel('Adopción (%)')
plt.ylabel('Crecimiento Estimado (%)')
plt.grid(True)
plt.show()
```
