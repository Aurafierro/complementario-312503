```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el archivo CSV directamente
archivo_csv = "LenguajesDesarrolloMovil.csv"

df = pd.read_csv("LenguajesDesarrolloMovil.csv")

#   columnas para manejarlas fácilmente
df.columns = ["Año", "Lenguaje", "Uso (%)", "Rendimiento", "Facilidad de Aprendizaje", "Extra1", "Extra2", "Extra3", "Extra4"]

# Filtrar las columnas relevantes
df = df[["Año", "Lenguaje", "Uso (%)", "Rendimiento", "Facilidad de Aprendizaje"]]

# Convertir las columnas numéricas
df["Uso (%)"] = df["Uso (%)"].astype(float)
df["Rendimiento"] = df["Rendimiento"].astype(float)
df["Facilidad de Aprendizaje"] = df["Facilidad de Aprendizaje"].astype(float)

# Estilo de gráficos
sns.set_style("whitegrid")
plt.figure(figsize=(16, 12))

# Gráfico 1: Correlación entre Uso y Rendimiento
plt.subplot(2, 2, 1)
sns.regplot(data=df, x="Uso (%)", y="Rendimiento", scatter=True, line_kws={"color": "blue"}, scatter_kws={"color": "skyblue", "alpha": 0.8})
plt.title("Correlación: Uso (%) vs Rendimiento", fontsize=14)
plt.xlabel("Uso (%)", fontsize=12)
plt.ylabel("Rendimiento", fontsize=12)

# Gráfico 2: Correlación entre Uso y Facilidad de Aprendizaje
plt.subplot(2, 2, 2)
sns.regplot(data=df, x="Uso (%)", y="Facilidad de Aprendizaje", scatter=True, line_kws={"color": "green"}, scatter_kws={"color": "lightgreen", "alpha": 0.8})
plt.title("Correlación: Uso (%) vs Facilidad de Aprendizaje", fontsize=14)
plt.xlabel("Uso (%)", fontsize=12)
plt.ylabel("Facilidad de Aprendizaje", fontsize=12)

# Gráfico 3: Correlación entre Rendimiento y Facilidad de Aprendizaje
plt.subplot(2, 2, 3)
sns.regplot(data=df, x="Rendimiento", y="Facilidad de Aprendizaje", scatter=True, line_kws={"color": "purple"}, scatter_kws={"color": "orchid", "alpha": 0.8})
plt.title("Correlación: Rendimiento vs Facilidad de Aprendizaje", fontsize=14)
plt.xlabel("Rendimiento", fontsize=12)
plt.ylabel("Facilidad de Aprendizaje", fontsize=12)

# Gráfico 4: Evolución del Uso (%) a lo largo de los años
plt.subplot(2, 2, 4)
sns.lineplot(data=df, x="Año", y="Uso (%)", hue="Lenguaje", marker="o", palette="tab10", linewidth=2.5)
plt.title("Evolución del Uso (%) por Lenguaje (2010-2024)", fontsize=14)
plt.xlabel("Año", fontsize=12)
plt.ylabel("Uso (%)", fontsize=12)
plt.legend(title="Lenguaje", loc="upper left")

# Ajustar diseño y mostrar los gráficos
plt.tight_layout()
plt.show()
```
