```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el archivo CSV directamente
archivo_csv = "LenguajesDesarrolloMovil.csv"

df = pd.read_csv("LenguajesDesarrolloMovil.csv")

#   columnas para manejarlas fcilmente
df.columns = ["Ao", "Lenguaje", "Uso (%)", "Rendimiento", "Facilidad de Aprendizaje", "Extra1", "Extra2", "Extra3", "Extra4"]

# Filtrar las columnas relevantes
df = df[["Ao", "Lenguaje", "Uso (%)", "Rendimiento", "Facilidad de Aprendizaje"]]

# Convertir las columnas numricas
df["Uso (%)"] = df["Uso (%)"].astype(float)
df["Rendimiento"] = df["Rendimiento"].astype(float)
df["Facilidad de Aprendizaje"] = df["Facilidad de Aprendizaje"].astype(float)

# Estilo de grficos
sns.set_style("whitegrid")
plt.figure(figsize=(16, 12))

# Grfico 1: Correlacin entre Uso y Rendimiento
plt.subplot(2, 2, 1)
sns.regplot(data=df, x="Uso (%)", y="Rendimiento", scatter=True, line_kws={"color": "blue"}, scatter_kws={"color": "skyblue", "alpha": 0.8})
plt.title("Correlacin: Uso (%) vs Rendimiento", fontsize=14)
plt.xlabel("Uso (%)", fontsize=12)
plt.ylabel("Rendimiento", fontsize=12)

# Grfico 2: Correlacin entre Uso y Facilidad de Aprendizaje
plt.subplot(2, 2, 2)
sns.regplot(data=df, x="Uso (%)", y="Facilidad de Aprendizaje", scatter=True, line_kws={"color": "green"}, scatter_kws={"color": "lightgreen", "alpha": 0.8})
plt.title("Correlacin: Uso (%) vs Facilidad de Aprendizaje", fontsize=14)
plt.xlabel("Uso (%)", fontsize=12)
plt.ylabel("Facilidad de Aprendizaje", fontsize=12)

# Grfico 3: Correlacin entre Rendimiento y Facilidad de Aprendizaje
plt.subplot(2, 2, 3)
sns.regplot(data=df, x="Rendimiento", y="Facilidad de Aprendizaje", scatter=True, line_kws={"color": "purple"}, scatter_kws={"color": "orchid", "alpha": 0.8})
plt.title("Correlacin: Rendimiento vs Facilidad de Aprendizaje", fontsize=14)
plt.xlabel("Rendimiento", fontsize=12)
plt.ylabel("Facilidad de Aprendizaje", fontsize=12)

# Grfico 4: Evolucin del Uso (%) a lo largo de los aos
plt.subplot(2, 2, 4)
sns.lineplot(data=df, x="Ao", y="Uso (%)", hue="Lenguaje", marker="o", palette="tab10", linewidth=2.5)
plt.title("Evolucin del Uso (%) por Lenguaje (2010-2024)", fontsize=14)
plt.xlabel("Ao", fontsize=12)
plt.ylabel("Uso (%)", fontsize=12)
plt.legend(title="Lenguaje", loc="upper left")

# Ajustar diseo y mostrar los grficos
plt.tight_layout()
plt.show()
```
