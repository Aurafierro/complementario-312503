```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el archivo CSV
file_path = 'FacilidadAprendizajeLenguajeProgramacion.csv'
data = pd.read_csv(file_path, encoding='ISO-8859-1')

# Limpiar las columnas de porcentaje convirtindolas a nmeros
data['Popularidad (%)'] = data['Popularidad (%)'].replace(r'%', '', regex=True).astype(float)
data['Facilidad de Aprendizaje (%)'] = data['Facilidad de Aprendizaje (%)'].replace(r'%', '', regex=True).astype(float)
data['Comunidad (%)'] = data['Comunidad (%)'].replace(r'%', '', regex=True).astype(float)
data['Uso en Automatizacizacion (%)'] = data['Uso en Automatizacizacion (%)'].replace(r'%', '', regex=True).astype(float)

# Eliminar filas con valores NaN en las columnas relevantes
data_clean = data.dropna(subset=['Popularidad (%)', 'Facilidad de Aprendizaje (%)'])

# Crear el grfico de dispersin con diferentes marcadores para cada lenguaje
plt.figure(figsize=(10, 8))
sns.scatterplot(data=data_clean, x='Popularidad (%)', y='Facilidad de Aprendizaje (%)', hue='Lenguaje', style='Lenguaje', s=100)

# Agregar etiquetas y ttulo
plt.title('Correlacin entre Popularidad y Facilidad de Aprendizaje de Lenguajes de Programacin', fontsize=14)
plt.xlabel('Popularidad (%)', fontsize=12)
plt.ylabel('Facilidad de Aprendizaje (%)', fontsize=12)

# Agregar la cuadrcula al fondo
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Mostrar el grfico
plt.legend(title='Lenguaje', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
