```python
# Importar las librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



# Nombre del archivo
file_name = 'SeguridadinformáticaCiberseguridad.csv'

# Leer el archivo con el separador adecuado y codificación
data = pd.read_csv(file_name, encoding='latin1', sep=';')

# Limpiar las columnas numéricas
columns_to_clean = [
    'Porcentaje de Uso 2023',
    'Porcentaje de Crecimiento 2023',
    'Comparación con 2022 (%)',
    'Porcentaje de Adopción 2024',
    'Crecimiento Esperado 2024 (%)',
]

# Reemplazar comas y convertir a valores numéricos
for col in columns_to_clean:
    data[col] = data[col].str.replace(',', '.').str.rstrip('%').astype(float)

# Seleccionar dos columnas para el gráfico
x_col = 'Porcentaje de Uso 2023'
y_col = 'Porcentaje de Crecimiento 2023'

# Crear el gráfico de dispersión con línea de tendencia
plt.figure(figsize=(8, 6))
sns.regplot(x=data[x_col], y=data[y_col], line_kws={"color": "red"})
plt.title(f'Relación entre {x_col} y {y_col}')
plt.xlabel(x_col)
plt.ylabel(y_col)
plt.grid(True)
plt.show()
```
