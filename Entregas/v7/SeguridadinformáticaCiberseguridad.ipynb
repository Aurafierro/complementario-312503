```python
# Importar las libreras necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



# Nombre del archivo
file_name = 'SeguridadinformticaCiberseguridad.csv'

# Leer el archivo con el separador adecuado y codificacin
data = pd.read_csv(file_name, encoding='latin1', sep=';')

# Limpiar las columnas numricas
columns_to_clean = [
    'Porcentaje de Uso 2023',
    'Porcentaje de Crecimiento 2023',
    'Comparacin con 2022 (%)',
    'Porcentaje de Adopcin 2024',
    'Crecimiento Esperado 2024 (%)',
]

# Reemplazar comas y convertir a valores numricos
for col in columns_to_clean:
    data[col] = data[col].str.replace(',', '.').str.rstrip('%').astype(float)

# Seleccionar dos columnas para el grfico
x_col = 'Porcentaje de Uso 2023'
y_col = 'Porcentaje de Crecimiento 2023'

# Crear el grfico de dispersin con lnea de tendencia
plt.figure(figsize=(8, 6))
sns.regplot(x=data[x_col], y=data[y_col], line_kws={"color": "red"})
plt.title(f'Relacin entre {x_col} y {y_col}')
plt.xlabel(x_col)
plt.ylabel(y_col)
plt.grid(True)
plt.show()
```
