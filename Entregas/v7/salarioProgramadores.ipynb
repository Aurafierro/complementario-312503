```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar los datos del archivo CSV con el delimitador correcto
archivo_csv = "salarioProgramadores.csv"
df = pd.read_csv(archivo_csv, sep=';')  # Especificar el delimitador

# Visualizar las primeras filas del DataFrame
print(df.head())

# Crear el diagrama de dispersión para correlación
plt.figure(figsize=(10, 6))
sns.regplot(x='Rating', y='Salary (INR)', data=df, scatter_kws={'alpha':0.6}, line_kws={"color":"red"})
plt.title('Diagrama de Correlación entre Rating y Salario', fontsize=16)
plt.xlabel('Rating', fontsize=12)
plt.ylabel('Salary (INR)', fontsize=12)
plt.grid(True)

# Mostrar el gráfico
plt.show()
```
